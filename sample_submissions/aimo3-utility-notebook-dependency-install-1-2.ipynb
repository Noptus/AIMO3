# AIMO3 Dependency Installation

Due to the AIMO3 competition not having internet access during the submissions, it is **not possible to use pip install in that notebook directly**. Furthermore, due to the need to restart the Kernel to install packages at the correct version, using a dataset of prebuilt wheels is also not possible.

The third option, which we implement in this complementary notebook, is to **pre-install the dependencies in a 'utility notebook'** into the `/kaggle/working` directory and then link it to the submission notebook as *utility notebook* via the `Add Input` button. To do this, click on `File` and `Set as Utility Script`.

Below are a few installs that pin exact dependencies in the Kaggle environment. It is necessary to uninstall a few packages first to avoid conflicts with newer vLLM versions. (You may not use vLLM, so in your case you may need to uninstall a different set of packages.)

```python
%pip uninstall --yes "tensorflow" "mxatplotlib" "keras" "scikit-learn" "protobuf" "numpy" "torch"
```

Below, we install all of the dependencies at the version we want to use for our submission.
Ignore the conflicts that appear in red when executing the cell below. This comes from the fact that the Kaggle environment (that can only be changed by Kaggle) contains many more packages with older versions, which cause these conflicts -- but they won't prevent you from successfully running this notebook.

```python
%pip install --target=/kaggle/working \
numpy==2.2.0 \
torch==2.8.0 \
vllm==0.10.2 \
pandas \
polars
```

The `Output` section on the right should now contain a long list of downloaded packages if you expand it, just below 10GB. (Hit the refresh button if you don't see anything.)
